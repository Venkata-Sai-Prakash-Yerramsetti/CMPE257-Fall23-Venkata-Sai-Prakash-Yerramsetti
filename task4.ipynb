```python
import pandas as pd
import numpy as np


from google.colab import drive
drive.mount('/content/drive/')

digits_data = pd.read_csv("/content/drive/MyDrive/datasets/ZipDigits.csv")

digits_data
```

```python
import pandas as pd

main_data = pd.DataFrame(digits_data)

# Define a function to split each row based on the first number
def divide_num(row):
    split_vals = row[0].split(maxsplit=1)  # Split only at the first space
    return pd.Series(split_vals)

# Apply the divide_num function to each row in the DataFrame
split_main_data = main_data.apply(divide_num, axis=1)

# Rename the columns in the split DataFrame
split_main_data.columns = ['y', 'x']
split_main_data['y'] = split_main_data['y'].apply(lambda x: round(float(x)))
# Display the split DataFrame
print(split_main_data)
```

```python
required_vals = [1, 5]

# Use boolean indexing to filter the DataFrame
required_df = split_main_data[split_main_data['y'].isin(required_vals)]
required_df['y'] =required_df['y'].astype(str)

# Replace values in the 'y' column with -1 if they contain '5'
required_df['y'] = required_df['y'].apply(lambda x: '-1' if '5' in x else x)
required_df['y'] = required_df['y'].apply(lambda x: round(float(x)))


# Display the filtered DataFrame
print(required_df)

def intense(cell):
    vals = [float(value) for value in cell.split()]
    return sum(vals) / len(vals)  # You can change this to median or other statistic

# Split the 'x' column into cells and calculate intensity
required_df['intensity'] = required_df['x'].apply(intense)

# Display the DataFrame with intensity values
print(required_df[['x', 'intensity']])

def sym(cell):
    vals = [float(value) for value in cell.split()]
    symmetry = sum(abs(vals[i] - vals[i + 1]) for i in range(len(vals) - 1)) / len(vals)
    return symmetry

# Split the 'x' column into cells and calculate symmetry
required_df['symmetry'] = required_df['x'].apply(sym)

# Display the DataFrame with symmetry values
print(required_df[['x', 'intensity' ,'symmetry' ,'y']])
```

```python
import matplotlib.pyplot as plt
category_1 = required_df[required_df['y'] == 1]
category_5 = required_df[required_df['y'] == -1]
plt.figure(figsize=(9, 7))
plt.scatter(category_1['intensity'], category_1['symmetry'], label='Class 1', color='blue')
plt.scatter(category_5['intensity'], category_5['symmetry'], label='Class 5', color='red')
plt.xlabel('Intensity')
plt.ylabel('Symmetry')
plt.title('Intensity vs. Symmetry')
plt.show()
```
